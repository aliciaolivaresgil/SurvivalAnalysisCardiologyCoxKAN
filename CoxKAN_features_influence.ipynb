```python
import pickle as pk
import pandas as pd
import matplotlib.pyplot as plt
import warnings
import numpy as np

from coxkan import CoxKAN

X = pk.load(open(f'data/X_admissions_general_pp=(ohe_norm)_cf_pp=(bool).df', 'rb'))
y = pk.load(open("data/y_admissions.df", "rb"))
tuned_params = pk.load(open(f'results/tuned_params_CoxKAN.pk', 'rb'))
scores = pk.load(open(f'results/scores_CoxKAN.pk', 'rb'))
```

# Get best hyperparams 

```python
cindexes = scores['cindex']

split = np.argmax(cindexes)

print(f'Best c-index in split {split}: {cindexes[split]}') 

print(f'Best hyperparams: {tuned_params[split]}')

num_hidden = tuned_params[split]['num_hidden']
hidden_dim = tuned_params[split]['hidden_dim']
base_fun = tuned_params[split]['base_fun']
grid = tuned_params[split]['grid']
noise_scale = tuned_params[split]['noise_scale']
lr = tuned_params[split]['lr']
steps = tuned_params[split]['steps']
lamb = tuned_params[split]['lamb']
lamb_entropy = tuned_params[split]['lamb_entropy']
lamb_coef = tuned_params[split]['lamb_coef']
```

# Train CoxKAN with all instances and best hyperparams

```python
df_full = pd.concat([X.reset_index(drop=True), y.reset_index(drop=True)], axis=1, join='outer').astype('float32')

if num_hidden == 0: 
    width = [len(df_full.columns)-2, 1]
else: 
    width = [len(df_full.columns)-2, hidden_dim, 1]

ckan = CoxKAN(width=width, base_fun=base_fun, grid=grid, noise_scale=noise_scale, seed=0)

ckan.train(df_full, duration_col='t', event_col='Ingreso', lr=lr, steps=steps, lamb=lamb, 
           lamb_entropy=lamb_entropy, lamb_coef=lamb_coef)

fig = ckan.plot(beta=20, scale=5)
plt.show()
```

## Linear symbolic fitting

```python
symbolic_lib_linear = dict()
symbolic_lib_linear['x'] = SYMBOLIC_LIB['x']

symbolic_fitted = []
for i, cov in enumerate(ckan.covariates): 
    if i not in pruned: 
        fun_name, fun, r2 = ckan.suggest_symbolic(l=0, i=i, j=0, lib=symbolic_lib_linear, verbose=False)
        if r2>0.99: 
            print(f'{cov}:\t {fun_name}\t {r2}')
            symbolic_fitted.append(i)
            r2 = ckan.fix_symbolic(l=0, i=i, j=0, fun_name=fun_name, verbose=False)

print(f'Fitted {len(symbolic_fitted)} from {len(covariates)}.')
```

## Polynomic symbolic fitting

```python
symbolic_lib_polynomic = dict()
symbolic_lib_polynomic['x^2'] = SYMBOLIC_LIB['x^2']
symbolic_lib_polynomic['x^3'] = SYMBOLIC_LIB['x^3']
symbolic_lib_polynomic['x^4'] = SYMBOLIC_LIB['x^4']
symbolic_lib_polynomic['1/x'] = SYMBOLIC_LIB['1/x']
symbolic_lib_polynomic['1/x^2'] = SYMBOLIC_LIB['1/x^2']
symbolic_lib_polynomic['1/x^3'] = SYMBOLIC_LIB['1/x^3']
symbolic_lib_polynomic['1/x^4'] = SYMBOLIC_LIB['1/x^4']
symbolic_lib_polynomic['sqrt'] = SYMBOLIC_LIB['sqrt']
symbolic_lib_polynomic['1/sqrt(x)'] = SYMBOLIC_LIB['1/sqrt(x)']
symbolic_lib_polynomic['exp'] = SYMBOLIC_LIB['exp']
symbolic_lib_polynomic['log'] = SYMBOLIC_LIB['log']
symbolic_lib_polynomic['sigmoid'] = SYMBOLIC_LIB['sigmoid']

for i, cov in enumerate(ckan.covariates): 
    if i not in pruned and i not in symbolic_fitted: 
        fun_name, fun, r2 = ckan.suggest_symbolic(l=0, i=i, j=0, lib=symbolic_lib_polynomic, verbose=False)
        if r2>0.99: 
            print(f'{cov}:\t {fun_name}\t {r2}')
            symbolic_fitted.append(i)
            r2 = ckan.fix_symbolic(l=0, i=i, j=0, fun_name=fun_name, verbose=False)

print(f'Fitted {len(symbolic_fitted)} from {len(covariates)}.')
```

## Trigonometric symbolic fitting

```python
symbolic_lib_trigonometric = dict()
symbolic_lib_trigonometric['sin'] = SYMBOLIC_LIB['sin']
symbolic_lib_trigonometric['tan'] = SYMBOLIC_LIB['tan']
symbolic_lib_trigonometric['tanh'] = SYMBOLIC_LIB['tanh']
symbolic_lib_trigonometric['sgn'] = SYMBOLIC_LIB['sgn']
symbolic_lib_trigonometric['arctan'] = SYMBOLIC_LIB['arctan']
symbolic_lib_trigonometric['arctanh'] = SYMBOLIC_LIB['arctanh']
symbolic_lib_trigonometric['cosh'] = SYMBOLIC_LIB['cosh']

for i, cov in enumerate(ckan.covariates): 
    if i not in pruned and i not in symbolic_fitted: 
        fun_name, fun, r2 = ckan.suggest_symbolic(l=0, i=i, j=0, lib=symbolic_lib_polynomic, verbose=False)
        if r2>0.99: 
            print(f'{cov}:\t {fun_name}\t {r2}')
            symbolic_fitted.append(i)
            ckan.fix_symbolic(l=0, i=i, j=0, fun_name=fun_name, verbose=False)

print(f'Fitted {len(symbolic_fitted)} from {len(covariates)}.')
```

```python
from PyPDF2 import PdfReader, PdfWriter, PdfMerger, PageObject, Transformation
from PyPDF2.generic import RectangleObject

def pdf_crop(pdf_path, upper=0, lower=0, left=0, right=0, output_file_name=None): 

    if upper<0 or upper>1 or lower<0 or lower>1 or left<0 or left>1 or right<0 or right>1: 
        raise Exception('lower, upper, left and right must be numerical values between 0 and 1.')
    
    #read pdf page
    reader = PdfReader(open(pdf_path, 'rb'))
    page = reader.pages[0]
    
    #calculate new margins 
    _, y = page.cropbox.upper_left
    x,_ = page.cropbox.lower_right
    upper = upper*float(page.mediabox.height)
    lower = lower*float(page.mediabox.height)
    left = left*float(page.mediabox.width)
    right = right*float(page.mediabox.width)
    
    #apply new margins 
    page.add_transformation(Transformation().translate(-left, -lower))
    new_mediabox = RectangleObject((0, 0, float(page.mediabox.width)-right-left, float(page.mediabox.height)-upper-lower))
    new_trimbox = RectangleObject((0, 0, float(page.mediabox.width)-right-left, float(page.mediabox.height)-upper-lower))
    page.mediabox = new_mediabox
    page.trimbox = new_trimbox
    new_upper_left = (0, float(y)-upper-lower)
    new_lower_right = (float(x)-right-left, 0)
    page.cropbox.upper_left = new_upper_left
    page.cropbox.lower_right = new_lower_right

    #save new pdf
    writer = PdfWriter()
    writer.add_page(page)
    if output_file_name: 
        file_name = output_file_name
    else: 
        file_name = pdf_path
    with open(file_name, 'wb') as f:
        writer.write(f)
    

def pdf_combine(pdf_list, orientation='horizontal', output_file_name='merge.pdf'):
    
    pages = []
    final_width = 0
    final_height = 0
    
    for pdf in pdf_list:
        reader = PdfReader(open(pdf, 'rb'))
        page = reader.pages[0]
        pages.append(page)
        if orientation == 'horizontal': 
            final_width+= float(page.mediabox.width)
            final_height = page.mediabox.height
        else: 
            final_width = page.mediabox.width
            final_height+= float(page.mediabox.height)
        
    final_page = PageObject.create_blank_page(None, final_width, final_height)
    
    for i_page in range(len(pages)-1, -1, -1): #pages in reverse order
        
        final_page.merge_page(pages[i_page])

        if i_page != 0: 
            if orientation == 'horizontal': 
                tx = float(pages[i_page-1].mediabox.width)
                ty = 0
            else: 
                tx = 0
                ty = float(pages[i_page-1].mediabox.height)
            final_page.add_transformation(Transformation().translate(tx, ty))
    
    writer = PdfWriter()
    writer.add_page(final_page)
    with open(output_file_name, 'wb') as f:
        writer.write(f)
```

```python
covs = ['HB', 'Urea', 'proBNP', 'K', 'Creatinina', 'Urato']
pdfs = [f'figs/comparative_{cov}.pdf' for cov in covs]
pdf_combine(pdfs, orientation='vertical', output_file_name='figs/comparative_bloodtest.pdf')

covs = ['FC_x', 'AnchuraQRS', 'FEVI']
pdfs = [f'figs/comparative_{cov}.pdf' for cov in covs]
pdf_combine(pdfs, orientation='vertical', output_file_name='figs/comparative_ecg_eco.pdf')

covs = ['SaO2', 'PAS', 'IMC', 'FC_y']
pdfs = [f'figs/comparative_{cov}.pdf' for cov in covs]
pdf_combine(pdfs, orientation='vertical', output_file_name='figs/comparative_other.pdf')
```
